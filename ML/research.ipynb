### EDA on products datasets from kaggle

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
amz_categories = pd.read_csv('amazon_categories.csv')

amz_categories = pd.DataFrame(amz_categories)

amz_categories
```

```python
amz_products = pd.read_csv('amazon_products.csv')

amz_products = pd.DataFrame(amz_products)

amz_products
```

```python
amz_products.describe()
```

```python
amz_products.info()
```

```python
amz = pd.merge(amz_categories,amz_products, left_on='id', right_on='category_id')

amz.head(10)
```

```python
amz = amz.drop(columns=['category_id','asin','imgUrl','productURL'])
amz.head(10)
```
